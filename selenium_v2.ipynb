```python
driver = webdriver.Chrome(service=service, options=chrome_options)

# driver.get('https://darmas.de/fahrzeugmarkt/details/6737593ce9b9ca11224fcf32')
driver.get('https://darmas.de/fahrzeugmarkt/details/67375b9ce9b9ca112257b30c')

details_elements = driver.find_elements(By.CSS_SELECTOR, "div.card-body")
combined_text_parts = []
for detail in details_elements[:5]:
    text = detail.text.strip()
    if text:
        combined_text_parts.append(text)
# Combine all parts into one block of text
combined_text = "\n".join(combined_text_parts)

# --- DEFINE THE STANDARDIZED HEADINGS ---
headings = [
    "Innenausstattung", "Lenkrad", "Räder & Fahrwerk",
    "Multimedia", "Assistenzsysteme", "Einparkhilfe",
    "Sitze", "Extras", "Sicherheit & Umwelt",
    "Licht & Scheinwerfer", "Airbags", "Emissionen", "Verbrauch & Umwelt", "Eigenschaften", "Kosten Modell"
]

# --- SPLIT THE COMBINED TEXT INTO SECTIONS BASED ON THE HEADINGS ---
# This regex looks for any of the headings followed by a newline and then captures text
# until it sees the next heading or the end of the string.
pattern = (
    r"(" + "|".join(re.escape(h) for h in headings) + r")\s*\n"  # Capture a heading followed by optional whitespace/newlines
    r"([\s\S]*?)"                                               # Non-greedy capture of the following content
    r"(?=(" + "|".join(re.escape(h) for h in headings) + r")|\Z)" # Stop at the next heading or end-of-string
)

matches = re.findall(pattern, combined_text, re.IGNORECASE)

classified_data = {}
for match in matches:
    # match[0] is the heading, match[1] is the content following the heading.
    heading = match[0].strip()
    content = match[1].strip()
    classified_data[heading] = content
    
# Click the "Ausstattung" button/tab.
ausstattung_tab = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.XPATH, "//ul[@class='nav nav-tabs']//a[normalize-space(text())='Ausstattung']"))
)
ausstattung_tab.click()

ausstattungs_elements = WebDriverWait(driver, 10).until(
    EC.visibility_of_all_elements_located((By.CSS_SELECTOR, ".col.text-truncate"))
)

# Extract and combine the text from the "Ausstattung" elements.
ausstattung_texts = []
for element in ausstattungs_elements:
    text = element.text.strip()
    if text:
        ausstattung_texts.append(text)
# Combine into one string, separated by commas (or adjust the separator as needed)
ausstattung_combined = ", ".join(ausstattung_texts)

classified_data["Ausstattung"] = ausstattung_combined
```

```python
# --- Phase 2: Process the Saved Raw Data ---

def process_raw_record(raw_record, headings):
    """
    Processes a raw record by classifying the combined text using
    the standardized headings and adding the 'Ausstattung' information.
    """
    combined_text = raw_record.get("combined_text", "")
    # Build regex to capture sections based on the standardized headings
    pattern = (
        r"(" + "|".join(re.escape(h) for h in headings) + r")\s*\n"
        r"([\s\S]*?)"
        r"(?=(" + "|".join(re.escape(h) for h in headings) + r")|\Z)"
    )
    matches = re.findall(pattern, combined_text, re.IGNORECASE)

    classified_data = {}
    for match in matches:
        heading = match[0].strip()
        content = match[1].strip()
        classified_data[heading] = content

    # Process "Ausstattung" from the raw list (if available)
    ausstattung_raw = raw_record.get("ausstattung_raw", [])
    if ausstattung_raw:
        # Join multiple entries with a comma.
        classified_data["Ausstattung"] = ", ".join(ausstattung_raw)

    # Retain the URL for reference
    classified_data["url"] = raw_record.get("url")
    
    return classified_data

# PHASE 2: Process the saved raw data into structured (classified) data.
processed_records = []
with open(raw_output_file, "r", encoding="utf-8") as fin, \
     open(processed_output_file, "w", encoding="utf-8") as fout:
    for line in fin:
        raw_record = json.loads(line)
        processed = process_raw_record(raw_record, headings)
        processed_records.append(processed)
        fout.write(json.dumps(processed, ensure_ascii=False) + "\n")
        
print("Processing phase complete. Processed data saved to", processed_output_file)

# --- Convert Processed Data to DataFrame and Save as CSV ---

df = pd.DataFrame(processed_records)
output_csv = "output/processed_data.csv"
df.to_csv(output_csv, index=False, encoding="utf-8")
print("CSV file saved to", output_csv)
```
