```python
### HYUNDAI
driver = webdriver.Chrome(service=service, options=chrome_options)

url = "https://www.hyundai.com/de/de/modelle.html"
driver.get(url)

# Optional: Wait until a key element is loaded
wait = WebDriverWait(driver, 3)
wait.until(EC.presence_of_element_located((By.TAG_NAME, "body")))

model_containers = driver.find_elements(By.CSS_SELECTOR, "div.rc-model-filter__item--text-wrapper")

data = []

for container in model_containers:
    # Extract model name
    try:
        model_name = container.find_element(By.CSS_SELECTOR, "h3.text__headline-main").text.strip()
    except Exception:
        model_name = "N/A"
    
    # Extract price details; there might be more than one price per model
    price_elements = container.find_elements(By.CSS_SELECTOR, "span.link-picture-text__price-number")
    prices = []
    for price_elem in price_elements:
        # Optionally remove any <sup> elements appended to the price
        try:
            sup = price_elem.find_element(By.TAG_NAME, "sup")
            driver.execute_script("arguments[0].remove();", sup)
        except Exception:
            pass
        prices.append(price_elem.text.strip())
    
    # Assign prices: assume the first price is the full price and the second (if available) is the leasing price
    full_price = prices[0] if len(prices) > 0 else "N/A"
    leasing_price = prices[1] if len(prices) > 1 else "N/A"
    
    # Append the row data to our list
    data.append({
        "date": datetime.datetime.now(),
        "website": url,
        "model": model_name,
        "full price": full_price,
        "leasing price": leasing_price
    })

# Create a DataFrame from the data list
df = pd.DataFrame(data)
# df["date"] = pd.to_datetime(df["date"])

driver.quit()
```

```python
# Point to your chromedriver path
driver = webdriver.Chrome(service=service, options=chrome_options)

# Open the Euler Group car listings page
url = "https://www.bmw.de/de/neufahrzeuge.html"
driver.get(url)

# Optional: Wait until a key element is loaded
wait = WebDriverWait(driver, 3)
wait.until(EC.presence_of_element_located((By.TAG_NAME, "body")))

all_text = driver.find_element(By.TAG_NAME, "body").text
print(all_text)
```

```python
# Extract all prices: look for the literal "Ab" followed by a price pattern (digits, dot, comma and €)
price_pattern = r'Ab\s*([\d\.,]+\s*€)'
prices = re.findall(price_pattern, all_text)

# Extract brand name occurrences. For this text, we look for "BMW" as a whole word.
brand_pattern = r'\bBMW\b'
brands = re.findall(brand_pattern, all_text)

# Remove duplicates from brands (if needed)
unique_brands = set(brands)

print("Prices found:")
for price in prices:
    print(price.strip())

print("\nBrand names found:")
for brand in unique_brands:
    print(brand)
```

```python
from bs4 import BeautifulSoup

def extract_car_models_and_prices_v3(file_path):
    # Open and read the HTML file
    with open(file_path, 'r', encoding='utf-8') as file:
        html_content = file.read()

    # Parse the HTML using BeautifulSoup
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find all car listing containers (assumed to be <article> tags with class "card-compact")
    listings = soup.find_all('article', class_='card-compact')

    # Iterate over each listing and try to extract the model and price
    for listing in listings:
        # Try to find the element containing the car model
        model_tag = listing.find(class_='card-compact__title')
        # Try to find the element containing the price
        price_tag = listing.find(class_='card-compact__price')

        model = model_tag.get_text(strip=True) if model_tag else "Model not found"
        price = price_tag.get_text(strip=True) if price_tag else "Price not found"

        print(f"Car Model: {model}, Price: {price}")

if __name__ == '__main__':
    extract_car_models_and_prices_v3(path+'output/raw_page_v3.txt')
```

```python
from bs4 import BeautifulSoup
import re

# Open and read the HTML file (raw_page.txt)
with open(path+"output/raw_page.txt", "r", encoding="utf-8") as file:
    html_content = file.read()

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(html_content, "html.parser")

# Find all car listing elements; based on the file, we expect them in <article> elements with class "card-compact"
car_cards = soup.find_all("article", class_="card-compact")
print(f"Found {len(car_cards)} car cards.")

# Loop through each car card and extract model name and price information
for card in car_cards:
    # Extract the model name from the <h3> element with class "card-compact__title"
    title_tag = card.find("h3", class_="card-compact__title")
    model_name = title_tag.get_text(strip=True) if title_tag else "N/A"
    
    # Extract the price text from an element with class "tile-button__subtitle"
    price_tag = card.find("div", class_="tile-button__subtitle")
    price_text = price_tag.get_text(strip=True) if price_tag else "N/A"
    
    # Optionally, extract just the numeric part of the price using regex
    price_value = None
    match = re.search(r'(\d[\d,\.]*)', price_text)
    if match:
        price_value = match.group(1).replace(",", "").replace(".", "")
    
    print("Model Name:", model_name)
    print("Price Text:", price_text)
    print("Numeric Price:", price_value)
    print("-" * 40)
```
