```python
def get_driver(service):
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument("--headless")
    chrome_options.add_argument("--disable-gpu")
    chrome_options.add_argument("--blink-settings=imagesEnabled=false")
    ua = random.choice(user_agents)
    chrome_options.add_argument(f"user-agent={ua}")
    chrome_options.add_experimental_option("excludeSwitches", ["enable-automation"])
    chrome_options.add_experimental_option("useAutomationExtension", False)
    # Explicitly set the page load strategy to "normal"
    chrome_options.page_load_strategy = "normal"
    
    driver = webdriver.Chrome(service=service, options=chrome_options)
    driver.execute_script("Object.defineProperty(navigator, 'webdriver', {get: () => undefined})")
    driver.implicitly_wait(10)
    return driver
```

```python
from playwright.async_api import async_playwright, expect
import asyncio
import sys # Import sys to check the platform

# The URL of the BMW configuration page
url = "https://configure.bmw.de/de_DE/configure/I20/61CF/FSDJL,P0300,S0230,S0248,S02PA,S02VB,S02VD,S02VH,S02VR,S0322,S03DN,S03HB,S03MF,S0413,S0420,S0428,S043M,S0494,S04AA,S04MA,S04NB,S04T2,S04U6,S04U9,S04V1,S0548,S0552,S05AC,S05AQ,S05AS,S05AT,S05AV,S05DM,S0654,S0688,S06AE,S06AF,S06AK,S06C4,S06NX,S06PA,S06U3,S0710,S0801,S0851,S0879,S08R3,S08R9,S08TF,S09T1,S09T2,S09TA,S09TB/SE000001"

async def run_bmw_click(): # Wrap logic in an async function
    async with async_playwright() as p:
        # Launch the browser (headless=False means you'll see the browser window)
        browser = await p.chromium.launch(headless=False) # Add await
        page = await browser.new_page() # Add await

        print(f"Navigating to {url}...")
        await page.goto(url, wait_until="load", timeout=90000) # Increased timeout slightly
        print("Page loaded.")

        # --- Handle Cookie Consent ---
        try:
            print("Looking for cookie consent banner...")
            accept_button = page.get_by_role("button", name="Alle akzeptieren")
            await accept_button.wait_for(state="visible", timeout=20000) # Increased timeout
            print("Cookie consent banner found. Clicking 'Alle akzeptieren'...")
            await accept_button.click() # Add await
            print("Cookie consent accepted.")
            await asyncio.sleep(1) # Use asyncio.sleep

        except Exception as e:
            print(f"Cookie consent banner not found or could not be clicked (maybe already accepted?): {e}")

        # --- Locate and Click the Target Button ---
        try:
            print("Locating 'Finanzieren & Leasen' button...")
            finance_button = page.get_by_role("button", name="Finanzieren & Leasen")

            print("Waiting for the button to be clickable...")
            await finance_button.wait_for(state="visible", timeout=25000) # Increased timeout
            await expect(finance_button).to_be_enabled(timeout=15000) # Increased timeout

            print("Clicking 'Finanzieren & Leasen'...")
            await finance_button.click() # Add await
            print("Button clicked successfully.")

            print("Waiting for 5 seconds to observe the result...")
            await asyncio.sleep(5) # Use asyncio.sleep

        except Exception as e:
            print(f"Error finding or clicking the 'Finanzieren & Leasen' button: {e}")
            await page.screenshot(path="error_screenshot_async.png") # Add await
            print("Screenshot saved as error_screenshot_async.png")

        finally:
            print("Closing browser.")
            await browser.close() # Add await

    print("Script finished.")

# --- How to Run ---

# Force SelectorEventLoop on Windows before starting Playwright
if sys.platform == "win32":
    print("Windows detected. Setting asyncio event loop policy...")
    asyncio.set_event_loop_policy(asyncio.WindowsSelectorEventLoopPolicy())
    print("Event loop policy set to WindowsSelectorEventLoopPolicy.")


# Option 1: If in Jupyter Notebook or IPython (with top-level await support)
# Just run this in a cell AFTER the code above:
# await run_bmw_click()

# Option 2: If running as a standard Python script (.py file)
# Define a main async function and use asyncio.run()
async def main():
   await run_bmw_click()

if __name__ == "__main__":
   # This asyncio.run() will now use the SelectorEventLoop on Windows
   # because we set the policy above.
   print("Starting script execution via asyncio.run()...")
   asyncio.run(main())
```
